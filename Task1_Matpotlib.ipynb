```python
china = df[df.country == 'China']
india = df[df.country == 'India']
us = df[df.country == 'United States']
indonesia = df[df.country == 'Indonesia']
Brazil = df[df.country == 'Brazil']
Nigeria = df[df.country == 'Nigeria']
```

```python
#fig = plt.figure()
#ax1 = plt.subplot2grid((1,1), (0,0))

style.use('ggplot')

plt.plot(china.year,china.population/10**6,label='China', linewidth=2)
plt.plot(india.year,india.population/10**6,label='India', linewidth=2)
plt.plot(us.year,us.population/10**6,label='US', linewidth=2)
plt.plot(us.year,indonesia.population/10**6,label='Indonesia', linewidth=2)
#plt.plot(us.year,Brazil.population/10**6)
plt.plot(us.year,Nigeria.population/10**6,label='Nigeria', linewidth=2)

#fig, (ax0, ax1) = plt.subplots(nrows=2)

plt.xlabel('Year')
plt.ylabel('Population Count')
plt.title('Top 5 Country Wise Comparion ')
#plt.legend(['China','India','US','Indonesia','Nigeria'])

#ax1.spines['left'].set_visible(True)
#ax1.spines['right'].set_visible(True)
#ax1.spines['top'].set_visible(True)
#ax1.spines['top'].set_visible(True)


plt.legend()
plt.grid(True,color='k')

plt.show() 
```
